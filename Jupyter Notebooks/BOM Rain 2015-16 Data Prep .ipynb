```r
library(readr)
library(dplyr)

file.list <- list.files(path = "C:/Users/lucin/Jupyter/2015-16/obs", pattern = "*.csv", full.names = T)

data2015.list <- lapply(file.list, read_csv)
data2015 <- bind_rows(data2015.list, .id = "id")
```

```r
stations <- read_csv("C:/Users/lucin/Jupyter/2015-16/spatial/StationData.csv")
```

Restrict observations to rainfall and QLD

```r
data2015_rain <- data2015[data2015$parameter == 'PRCP',]
```

```r
stations_qld <- stations[stations$REGION == 'QLD',]
station_list_qld <-  unique(stations_qld$station_number) 
data2015_rain_qld <- data2015_rain[data2015_rain$station_number %in% station_list_qld,]
```

```r
dim(data2015)
dim(data2015_rain)
dim(data2015_rain_qld)
```

Merge station data with observation data

```r
data2015_rain_qld_stations = merge(data2015_rain_qld, stations_qld, by="station_number", all.x=TRUE)
```

```r
head(data2015_rain_qld_stations)
```

```r
length(unique(data2015_rain_qld_stations$station_name ) )
unique(data2015_rain_qld_stations$station_name ) 
```

Convert UTC timestamp to local time. All observations are in QLD, so they will all be in the same timezone

```r
data2015_rain_qld_stations$valid_start <- as.POSIXct(data2015_rain_qld_stations$valid_start, origin='1970-01-01', tz="Australia/Brisbane")
data2015_rain_qld_stations$valid_end <- as.POSIXct(data2015_rain_qld_stations$valid_end, origin='1970-01-01', tz="Australia/Brisbane")
```

```r
head(data2015_rain_qld_stations)
summary(data2015_rain_qld_stations)
```

```r
data2015_rain_qld_stations[data2015_rain_qld_stations$value == 461.8,]
```

```r
with(data2015_rain_qld_stations, plot(valid_start, value))
```

Determine how many observations there are per station

```r
station_summary <- aggregate(data2015_rain_qld_stations$valid_start, list(data2015_rain_qld_stations$station_name), FUN=length)
station_summary <- station_summary[order(-station_summary$x),]
```

```r
station_summary
```

Add date field

```r
data2015_rain_qld_stations$day <- as.Date(data2015_rain_qld_stations$valid_start)
head(data2015_rain_qld_stations)
```

```r
daysummary <- aggregate(data2015_rain_qld_stations$value, list(data2015_rain_qld_stations$station_name,data2015_rain_qld_stations$day ), FUN=sum)
```

```r
SUNCOAST <- daysummary[daysummary$Group.1 == 'SUNSHINE COAST AIRPORT', ]
SUNCOAST

```
